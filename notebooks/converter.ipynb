```python
import pandas as pd, json
```

```python

df = pd.read_csv("nappeBerrechid.csv")

# Extract coordinates
coords = df[["x", "y"]].values.tolist()

# Ensure polygon closure
if coords[0] != coords[-1]:
    coords.append(coords[0])

# Build GeoJSON
geojson = {
    "type": "FeatureCollection",
    "features": [
        {
            "type": "Feature",
            "properties": {"name": "Nappe Berrechid"},
            "geometry": {"type": "Polygon", "coordinates": [coords]},
        }
    ],
}

# Save to file
with open("nappeBerrechid.geojson", "w") as f:
    json.dump(geojson, f)
```
